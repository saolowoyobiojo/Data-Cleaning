# Heights column
now we need to properly clean the height column, all metrics should be in cm 

```python
# spliting the rorws in order to deal with non numeric data
data2 = df[~df["Height (cm)"].str.isnumeric()] #the non numeric part
data1 = df[df["Height (cm)"].str.isnumeric()] #the numeric part
```

```python
#Converting the heights in data2 (feets) to cm, so that it look like that of data1
df["Height (cm)"]=df["Height (cm)"].str.replace("'"," ")
data2[["a","b"]]=data2["Height (cm)"].str.split("'", expand=True)
data2["b"]= data2["b"].str.extract(pat='(\d+)', expand=False)
new = ["a","b"]
for n in new:
  data2[n] = data2[n].astype("int")
data2["Height (cm)"] = (data2["a"]*30.48 + data2["b"]*2.54).round()
data2.drop(["a","b"], axis=1, inplace=True)
data2.head()
```

# Weight column
now we need to properly clean the weight column, all metrics should be in lbs 

```python
#converting the weight in data 1
data1["Weight (kg)"] = data1["Weight (kg)"].astype("int")
data1["Weight (kg)"] = (data1["Weight (kg)"]*2.20462).round()
#remove lbs 
data2["Weight (kg)"] = data2["Weight (kg)"].str.extract(pat='(\d+)', expand=False)
```

```python
data1.head(2)
```

```python
df = pd.concat([data1,data2]) #to join the rows we splitted back together 
```

```python
df.sample(2)
```

```python
#renaming Weight (kg) to Weight (lbs)
df.columns = df.columns.str.replace("Weight (kg)", "Weight (lbs)")
```
